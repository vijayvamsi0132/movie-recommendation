```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
movies['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
import re

def recommend(movie: str):
    # Escape user input to avoid regex special chars
    pattern = re.escape(movie)
    # Find titles that contain the substring (case-insensitive)
    candidates = movies[movies['title'].str.contains(pattern, case=False, na=False)]

    if candidates.empty:
        raise ValueError(f"No titles found containing: {movie!r}")

    # If multiple candidates, you can choose the first or ask the user to be more specific.
    # Here, we take the closest by substring length (shortest additional chars).
    # Alternatively, just pick the first: candidates.index[0]
    candidates = candidates.assign(extra_len=candidates['title'].str.len() - len(movie))
    movie_index = candidates.sort_values(['extra_len', 'title']).index[0]

    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    return [movies.iloc[i].title for i, _ in movies_list]
```

```python
recommend("legacy")
```
